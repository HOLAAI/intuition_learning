- vectorized implementation 에 대한 이해와 연습을 위주로 한다
- 여기서는 학습하려는 값을 wiki에 작성한 w로 표시 안하고 theta로 표기한다.

```python
import numpy as np   ## numpy import
```

```python
### 학습을 위해서 실제 맞는 값을 생성하는 함수
def real_h(X) :
    results = []
    for row in X :
        results.append(row[1] * 2 + row[2] * 3 + row[3] * 4 + row[4] * 5 + 2)
    return results
```

```python
### 학습데이터를 생성한다.
### x[0] is theta0,  bais 를 만들어야 하기 때문에 X[0] = 1
X = np.array([
    [1, 1, 2, 3, 4],
    [1, 5, 6, 7, 8],
    [1, 3, 2, 1, 2],
    [1, 5, 1, 1, 2]
])
```

```python
### 실제 값을 생성한다. (학습데이터의 정답값) 
y = real_h(X)
```

```python
### theta를 random 값으로 초기화
theta = np.random.rand(1, 5)
```

```python
### linear 함수로 가설 함수 구현
### 아래처럼 행렬의 내적으로 구할 필요는 없지만 이번 코드는 가급적 내적을 사용한다. (vectorized implementation 사용)
hypothesis = lambda  theta, X :  np.dot(theta, X.T)
```

- batch 방식으로 학습을 진행한다. 
- MSE cost function 을 편미분 하여 theta 값에 빼주면 학습이 된다.(Gradient Descent)
- 포멧은 아래와 같다.
- Θ = Θ - α(1/m)Σ(hypothesis(X) - y)x(i)

```python
learning_rate = 0.0250  ## α 값 
for i in range(1000) :
    # 예측값 생성
    Y_hat = hypothesis(theta, X)
    
    ## 실제값 - 예측값
    Y_diff= Y_hat - y
    
    ## Σ(hypothesis(X) - y)x(i)
    temp = np.dot(Y_diff, X)
    
    ## Θ = Θ - α(1/m)temp
    theta = theta -  learning_rate * (temp / X.shape[0])
    
    ### 여기서는  단순하게 확인을 위하여 MSE의 에러값을 출력 하는것이 아니라 차이값의 제곱을 화면에 출력하여 학습이 되고 있는지(오류가 줄어드는지) 확인
    ### 학습된 theta(Θ) 값을 같이 출력
    print("==> error", np.sum(Y_diff) ** 2, theta)
```
